```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
# interactive plots, must add before each plot or they will all add to same plot
```

```python
# MidEast January
# with xr.open_dataset('../MidEastJan.nc') as ds:  (on Mac)
with xr.open_dataset('../Oct30/MidEastJan2018.nc') as ds:
    print(ds)
```

```python
with xr.open_dataset('../Oct30/MidEastJan2018.nc') as ds:
    fig, (ax1,ax2,ax3) = plt.subplots(3, 1, sharex=True, figsize=(9,10))
    fig.subplots_adjust(hspace = 0.1)
    
    ax1.set_title('MidEast ADCP 150 kHz - January 2018 - u, v, w velocities')
    ax1.set_ylabel('Depth [m]')
    im1 = ax1.pcolormesh(np.linspace(0,30,720), -ds.depth, ds.u.T, rasterized=True, cmap='RdBu_r', vmin=-0.2, vmax=0.2)
    cbar1 = fig.colorbar(im1, ax=ax1, fraction=0.05, pad=0.02, aspect=40)
    cbar1.set_label('Velocity [m/s]')
    
    im2 = ax2.pcolormesh(np.linspace(0,30,720), -ds.depth, ds.v.T, rasterized=True, cmap='RdBu_r', vmin=-0.2, vmax=0.2)
    cbar2 = fig.colorbar(im2, ax=ax2, fraction=0.05, pad=0.02, aspect=40)
    
    ax3.set_xlabel('Time [days]')
    im3 = ax3.pcolormesh(np.linspace(0,30,720), -ds.depth, ds.w.T, rasterized=True, cmap='RdBu_r', vmin=-0.03, vmax=0.03)
    cbar3 = fig.colorbar(im3, ax=ax3, fraction=0.05, pad=0.02, aspect=40)
```
