```python
# plot histogram

fig,ax = plt.subplots(1,1,figsize=(4.5,3))

bins=[7,10,10,13,13,16,16,19]
colors=['orange','deepskyblue','yellowgreen','tomato']
alpha=1
ax.hist([fall,winter,spring,summer],rwidth=0.75,orientation='vertical',\
        color=colors,bins=bins,align='mid',alpha=alpha,bottom=0.1,\
        label=['Fall','Winter','Spring','Summer'],edgecolor='black')
ax.set_ylabel('# of events',fontsize=14)
ax.set_xlabel('Delay [days]',fontsize=14)
#ax.set_title('Deep response delay',fontsize=14)
ax.tick_params(labelsize=14)
ax.set_yticks([0.1,1.1,2.1,3.1,4.1,5.1,6.1])
ax.set_yticklabels(['0','1','2','3','4','5','6'])
ax.set_ylim(0,6.2)
ax.set_xticks([8.5,11.5,14.5,17.5])
ax.set_xticklabels(['7-9','10-12','13-15','16-18'])
ax.set_xlim(7,19)
fig.legend(loc=[0.37,0.74],fontsize=12,ncol=2,frameon=False)
fig.tight_layout()

plt.show()

plt.savefig(fname=f'./wind_histogram.pdf',format='pdf')
```
