# Data summary

Summary of the state of NetCDF velocity data for the Axis and Upper Slope ADCP nodes, with an emphasis on data quality during years 2013, 2014, 2017, and 2018.

# Issues

## Current issues and ONC requests

- New NetCDF data have updated correlation and error velocity *threshold* values in the metadata, but still do not contain the plottable data for correlation, intensity, or error velocity. We request this data be added to the NetCDF files. See section 3.2.
- For the new Axis 75 kHz data, both averaged and raw data show an increase in velocity magnitudes / variance, so whatever changes were made do make a difference. We request a record of changes made to *any* instrument at the Axis or Upper Slope sites, and why those changes were made. See section 4.

## Fixed issues

- Regarding the Axis 75 kHz correlation saturation issue, the ExcludeADCPbeamNumber fields on the ONC website show that the problem beams were turned off for the relevant instruments, for the relevant periods, so this issue appears to be accounted for. Device Action notes indicate the reason being the manufacturer identified faulty DSP boards for the culprit beams. In the new MAT data the saturation is still visible when plotting the correlation for each beam, but as the faulty beams are turned off it does not contribute to the data.

# Axis 75 kHz data comparison

Comparing new and previous versions of Axis 75 kHz velocity data.

## Single-day velocity comparison (*u* velocities; July 1, 2014), 15-minute averaged data 

- 'New 3-beam data' was downloaded on February 17, 2021, and includes updated metadata. However, there is a further difference in the magnitudes of the data (see below), so I am *assuming* that beam-1 was shut off for device 23065 that was swapped in May 2014 (this is listed in the ExcludeADCPbeamNumber field).
- The new averaged data shows increased velocity magnitudes, reflected in greater variance.
- 'Previous 3-beam data' was downloaded on January 27, 2021, and included beam-2 being shut off for all data beyond September 1, 2013. 

New 3-beam data (NetCDF), variance @ -800 m = 0.0005048 (m/s)^2:
![image](https://user-images.githubusercontent.com/57200139/108579782-60110880-72dd-11eb-9cdf-e3833bf9dddf.png)
Previous 3-beam data (NetCDF), variance @ -800 m = 0.0004001 (m/s)^2:
![image](https://user-images.githubusercontent.com/57200139/108579708-08729d00-72dd-11eb-8c10-c6f8cb5c2a64.png)

## Single-day velocity comparison (*u* velocities; July 1, 2014), raw 2-second data 

- The new raw data shows the same increase in velocity magnitudes, and is more visually apparent in these plots than for the averaged data - the new data is much darker than the previous data. I confirmed the magnitude increase in the new raw data in both NetCDF and MAT format. It is again reflected in increased variance.
