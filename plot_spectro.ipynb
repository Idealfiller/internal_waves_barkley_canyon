# Spectrogram plots

Loads depth-frequency-time data and then plots spectrograms for various circumstances.

## Imports
Necessary modules for analysis.

```python
# import modules

import xarray as xr
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.dates as pldates
import numpy as np
for i in range(2):
    %matplotlib notebook
```

```python
# import data

adcp = 'Slope'     # Slope(2013,2014,2017,2018), Axis75(2013,2014), Axis55(2017,2018)
year = 2013
ds_in = xr.open_dataset(f'../Data/data/spectra/spectra_{adcp}_{year}_0.nc')

n_seg = ds_in.n_seg
if n_seg > 1:
    ds = [ds_in]
    for i in range(n_seg):
        if i > 0:
            ds_temp = xr.open_dataset(f'../Data/data/spectra/spectra_{adcp}_{year}_{i}.nc')
            ds.append(ds_temp)
elif n_seg == 1:
    ds = [ds_in]

#print(ds)
```

```python
# extract plotting variables

t_stamp = int(ds[0].t_stamp)
start_date = dt.datetime(t_stamp,1,1)                                   # input start date in YYYY,MM,DD
end_date = dt.datetime(t_stamp+1,1,1)                                   # input end date in YYYY,MM,DD
depth = ds[0].depth.values
f = ds[0].specf_s.values
```

## Depth-specific PSD spectrograms (annual)

```python
# WKB scale and select depth

WKB_u,WKB_v = [],[]
for i in range(n_seg):
    WKB_u_temp = ds[i].Sxxu_d / ds[0].WKB_scale
    WKB_v_temp = ds[i].Sxxv_d / ds[0].WKB_scale
    WKB_u.append(WKB_u_temp)
    WKB_v.append(WKB_v_temp)
    
if adcp == 'Slope':
    didx = -9           # -9 for lower depth (-315), -24 for mid-depth (-195)
elif adcp == 'Axis75':
    didx = -7           # -7 for lower depth (-920), -20 for mid-depth (-800)
elif adcp == 'Axis55':
    didx = -4           # -4 for lower depth (-913), -10 for mid-depth (-793)
    
u,v = [],[]
for i in range(n_seg):
    u_temp = WKB_u[i][didx]
    v_temp = WKB_v[i][didx]
    u.append(u_temp)
    v.append(v_temp)
    
d_stamp = ds[0].depth[didx].values
```

```python
# 'whiten' spectrogram (flatten it) by multiplying Sxx by frequency**2

Sxxu_w,Sxxv_w = [],[]
for i in range(n_seg):
    Sxxu_white = np.copy(u[i])
    for j in range(len(ds[i].spect_s)):
        Sxxu_white[:,j] = Sxxu_white[:,j]*(f**2)
    Sxxv_white = np.copy(v[i])
    for j in range(len(ds[i].spect_s)):
        Sxxv_white[:,j] = Sxxv_white[:,j]*(f**2)
    Sxxu_w.append(Sxxu_white)
    Sxxv_w.append(Sxxv_white)
```

```python
# plot spectrogram

fig, (ax0,ax1) = plt.subplots(2, 1, figsize=(7.2,6), sharex=True, sharey=True)
fig.subplots_adjust(hspace=0.15)
fig.text(0.5, 0.94, f'-{d_stamp} m - {adcp} - {t_stamp}', ha='center', fontsize=14)
fig.text(0.02, 0.5, 'Frequency [Hz]', va='center', rotation='vertical', fontsize=14)
fig.text(0.96, 0.5, r'PSD*$\omega^2$ [(m/s)$^2$*Hz]', va='center', rotation='vertical', fontsize=14)
fig.text(0.5, 0.01, f'Time [months]', ha='center',fontsize=14)
fig.text(0.205, 0.886, 'WKB-scaled', ha='center', fontsize=14)
fig.text(0.79, 0.886, 'Whitened', ha='center', fontsize=14)

if adcp == 'Slope':
    fig.text(0.5, 0.886, 'Cross-slope', ha='center', fontsize=14)
    fig.text(0.5, 0.478, 'Along-slope', ha='center', fontsize=14)
    vmin = 1e-9
    vmax = 5e-7
elif adcp == 'Axis55' or adcp == 'Axis75':
    fig.text(0.5, 0.886, 'Cross-canyon', ha='center', fontsize=14)
    fig.text(0.5, 0.478, 'Along-canyon', ha='center', fontsize=14)
    vmin = 1e-9
    vmax = 5e-7

for i in range(n_seg):
    im0 = ax0.pcolormesh(ds[i].spect_s, f, Sxxu_w[i], rasterized=True, \
                        norm=colors.LogNorm(vmin=vmin, vmax=vmax), cmap='plasma',shading='auto')  
    im1 = ax1.pcolormesh(ds[i].spect_s, f, Sxxv_w[i], rasterized=True, \
                        norm=colors.LogNorm(vmin=vmin, vmax=vmax), cmap='plasma',shading='auto')  

cbar0 = fig.colorbar(im0, ax=ax0, fraction=0.05, pad=0.01, aspect=15, extend='both')
cbar1 = fig.colorbar(im1, ax=ax1, fraction=0.05, pad=0.01, aspect=15, extend='both')    
cbar0.ax.tick_params(labelsize=14)
cbar1.ax.tick_params(labelsize=14)

ax0.patch.set_facecolor('grey')
ax1.patch.set_facecolor('grey')

ax0.set_yscale('log')
ax0.set_ylim(5e-6,5.55e-4)
date_form = pldates.DateFormatter("%m")
ax0.xaxis.set_major_formatter(date_form)
ax0.set_xlim(start_date,end_date)
ax0.tick_params(labelsize=14)
ax1.tick_params(labelsize=14)

for i in [ax0,ax1]:
    pltcolor = 'white'
    bbox = dict(facecolor='black', alpha=0.25, boxstyle='square, pad=0.1', lw=0)
    i.axhline(2.236e-5, color = pltcolor,ls=(0,(1,3)),lw=2)
    i.axhline(1.732e-5, color = pltcolor,ls=(0,(1,3)),lw=2) 
    i.axhline(1.161e-5, color = pltcolor,ls=(0,(1,3)),lw=2) 
    i.axhline(4.026e-4, color = pltcolor,ls=(0,(1,3)),lw=2)
    i.text(0.01,0.08,'K$_1$',fontsize=14,color=pltcolor,transform=i.transAxes,bbox=bbox)
    i.text(0.01,0.18,'$f$',fontsize=14,color=pltcolor,transform=i.transAxes,bbox=bbox)
    i.text(0.01,0.355,'M$_2$',fontsize=14,color=pltcolor,transform=i.transAxes,bbox=bbox)
    i.text(0.01,0.835,'N',fontsize=14,color=pltcolor,transform=i.transAxes,bbox=bbox)

plt.show()

plt.savefig(fname=f'./plots/spectro_plots/spectro_psd_{adcp}_{d_stamp}_{t_stamp}.pdf',format='pdf')
```

## Depth-specific rotary spectrograms (annual)

```python
# WKB scale and select depth

WKB_cw,WKB_ccw = [],[]
for i in range(n_seg):
    WKB_cw_temp = ds[i].cw_d / ds[0].WKB_scale
    WKB_ccw_temp = ds[i].ccw_d / ds[0].WKB_scale
    WKB_cw.append(WKB_cw_temp)
    WKB_ccw.append(WKB_ccw_temp)
    
if adcp == 'Slope':
    didx = -9           # -9 for lower depth (-315), -24 for mid-depth (-195)
elif adcp == 'Axis75':
    didx = -7           # -7 for lower depth (-920), -20 for mid-depth (-800)
elif adcp == 'Axis55':
    didx = -4           # -4 for lower depth (-913), -10 for mid-depth (-793)    

cw,ccw = [],[]
for i in range(n_seg):
    cw_temp = WKB_cw[i][didx]
    ccw_temp = WKB_ccw[i][didx]
    cw.append(cw_temp)
    ccw.append(ccw_temp)
    
d_stamp = ds[0].depth[didx].values
```

```python
# 'whiten' spectrogram (flatten it) by multiplying Sxx by frequency**2

Sxxcw_w,Sxxccw_w = [],[]
for i in range(n_seg):
    Sxxcw_white = np.copy(cw[i])
    for j in range(len(ds[i].spect_s)):
        Sxxcw_white[:,j] = Sxxcw_white[:,j]*(f**2)
    Sxxccw_white = np.copy(ccw[i])
    for j in range(len(ds[i].spect_s)):
        Sxxccw_white[:,j] = Sxxccw_white[:,j]*(f**2)
    Sxxcw_w.append(Sxxcw_white)
    Sxxccw_w.append(Sxxccw_white)
```

```python
# plot spectrogram

fig, (ax0,ax1) = plt.subplots(2, 1, figsize=(7.2,6), sharex=True, sharey=True)
fig.subplots_adjust(hspace=0.15)
fig.text(0.5, 0.94, f'-{d_stamp} m - {adcp} - {t_stamp}', ha='center', fontsize=14)
fig.text(0.02, 0.5, 'Frequency [Hz]', va='center', rotation='vertical', fontsize=14)
fig.text(0.96, 0.5, r'Rotary PSD*$\omega^2$ [(m/s)$^2$*Hz]', va='center', rotation='vertical', fontsize=14)
fig.text(0.5, 0.01, f'Time [months]', ha='center',fontsize=14)
fig.text(0.205, 0.886, 'WKB-scaled', ha='center', fontsize=14)
fig.text(0.79, 0.886, 'Whitened', ha='center', fontsize=14)

if adcp == 'Slope':
    vmin = 1e-9
    vmax = 5e-7
elif adcp == 'Axis55' or adcp == 'Axis75':
    vmin = 1e-9
    vmax = 5e-7

fig.text(0.5, 0.884, 'CW', ha='center', fontsize=14)
fig.text(0.5, 0.473, 'CCW', ha='center', fontsize=14) 
    
for i in range(n_seg):
    im0 = ax0.pcolormesh(ds[i].spect_s, f, Sxxcw_w[i], rasterized=True, \
                        norm=colors.LogNorm(vmin=vmin, vmax=vmax), cmap='plasma',shading='auto')  
    im1 = ax1.pcolormesh(ds[i].spect_s, f, Sxxccw_w[i], rasterized=True, \
                        norm=colors.LogNorm(vmin=vmin, vmax=vmax), cmap='plasma',shading='auto')  

cbar0 = fig.colorbar(im0, ax=ax0, fraction=0.05, pad=0.01, aspect=15, extend='both')
cbar1 = fig.colorbar(im1, ax=ax1, fraction=0.05, pad=0.01, aspect=15, extend='both')    
cbar0.ax.tick_params(labelsize=14)
cbar1.ax.tick_params(labelsize=14)

ax0.patch.set_facecolor('grey')
ax1.patch.set_facecolor('grey')

ax0.set_yscale('log')
ax0.set_ylim(5e-6,5.55e-4)
date_form = pldates.DateFormatter("%m")
ax0.xaxis.set_major_formatter(date_form)
ax0.set_xlim(start_date,end_date)
ax0.tick_params(labelsize=14)
ax1.tick_params(labelsize=14)

for i in [ax0,ax1]:
    pltcolor = 'white'
    bbox = dict(facecolor='black', alpha=0.25, boxstyle='square, pad=0.1', lw=0)
    i.axhline(2.236e-5, color = pltcolor,ls=(0,(1,3)),lw=2)
    i.axhline(1.732e-5, color = pltcolor,ls=(0,(1,3)),lw=2) 
    i.axhline(1.161e-5, color = pltcolor,ls=(0,(1,3)),lw=2) 
    i.axhline(4.026e-4, color = pltcolor,ls=(0,(1,3)),lw=2)
    i.text(0.01,0.08,'K$_1$',fontsize=14,color=pltcolor,transform=i.transAxes,bbox=bbox)
    i.text(0.01,0.18,'$f$',fontsize=14,color=pltcolor,transform=i.transAxes,bbox=bbox)
    i.text(0.01,0.355,'M$_2$',fontsize=14,color=pltcolor,transform=i.transAxes,bbox=bbox)
    i.text(0.01,0.835,'N',fontsize=14,color=pltcolor,transform=i.transAxes,bbox=bbox)

plt.show()

plt.savefig(fname=f'./plots/spectro_plots/spectro_rot_{adcp}_{d_stamp}_{t_stamp}.pdf',format='pdf')
```
